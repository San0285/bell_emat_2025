## Analyzing Online Interest in Recent and Upcoming World of Warcraft Expansions

### Introduction and Research Motivation

World of Warcraft is an MMORPG (Massively multiplayer online role-playing game) that is in it's 21st year since it's release on November 23rd, 2004. Since then, World of Warcraft has launced 10 expansions, and all of these expansions have been met with a surge of community excitement, speculation and online activity. Players and content creators response to expansion reveals, cinematics, and feature announcements create hype across all social media platforms. With the upcoming release of their eleventh expansion, Midnight, in Spring of 2026, there has been a lot of buzz online, particularly due to the newest feature, player housing. My motivation for this researh is to explore the following question: </br>

*How does online interest in World of Warcraft expansions rise, peak, and evolve across different reveal and launch periods, and how does the upcoming expansion Midnight compare to Dragonflight and The War Within?*

To find the answer to this, I analyzed engagement trends using two data sources:
- Google Trends, which measures relative search interest
- YouTube Data API, which reflects community engagement through video views and creator activity.

My *hypothesis* was: </br>

Online interest in World of Warcraft expansions follows a cyclical pattern, peaking near official announcements, cinematic reveals, and launch windows.

### Data Sources and Structure

#### Google Trends </br>

Google Trends provides a weekly search interest based on a scale between 0-100. For each expansion- Dragonflight (2022), The War Within (2024), and Midnight (2025/2026) I collected search data using the terms: </br>
- *Dragonflight* - WoW Dragonflight, World of Warcraft Dragonflight, WoW Dragonflight 2022 </br>
- *The War Within* - WoW The War Within, World of Warcraft The War Within, WoW The War Withing 2024, and Worldsoul Saga </br>
- *Midnight* - WoW Midnight, World of Warcraft Midnight, player housing, and WoW Midnight 2026 </br>

Each dataset consisted of columns including the date, and key terms, and rows that represented one date and a search interest score under each key term. With this data I was able to compare:</br>
- peak interst scores
- interest trends over time
across multiple years (2022-2025)

#### YouTube Data API

YouTube Data API provided me data measuring community engagement, so that I could compare popularity across all three expansions, and track how interest evolved over time. The YouTube data included:
- Video title
- Video description
- Channel title
- Date Published
- Counts for views, likes, and comments

The dataset contained 20 videos per expansion, and showed a comparison of:
- Total views
- Median views
- standout high-performing videos

While Google Trends meausures initial curiosity, YouTube shows how the community continues to engage with and dicuss each expansion through long-form content.

### Analysis and Interpretation

#### Google Search Behavior

The three expansions followed a recognizable patter that I wasn't too surprised at:
- *Dragonflight* peaked instantly at 100, then declined rapidly
- *The War Within* built up slower but ultimately reached a peak of 78 during its announcement
- *Midnight*, althought its still to early, it showed a promising initial spike and a gradual settling pattern

Dragonflight had a stronger single peak, while The War Within displayed a broader sustained rise leading up to its peak interest. These results strongly support the fact that hype cycles are predictable.

#### YouTube Engagement

YouTubes tells a different story, but it still compliments Google Trends. 
- *The War Within* had the highest total views 23.2 million and the highest median views
- *Dragonflight* had large outlier videos, example, a cinematic with 8.6 million views. But,it had a lower median than The War Within
- *Midnight* still in its early phase, but its engagement so far has been pretty strong, especially for the housing and cinematic content

 This shows that The War Within resonated more with the community and creator involvement, even if its Google spike trend was smaller than Dragonflights.

### Conclusion

The data I collected clearly supports my hypothesis that, *Online interest in Worlf of Warcraft expansions follows cyclical patterns*. Rising sharply during announcements and declining afterwards. Across Google and Youtube I found that:
- *Dragonflight* genterated the strongest immediate reveal-day search interest. </br>
- *The War Within* had the most sustained creator engagement and the highest view totals on YouTube.</br>
- *Midnight* data is still building as it nears it's launch, but there is evidence of strong early interest, especially around player housing.</br>

Google Trends is refecting broad public curiosity, while YouTube is reflecting deeper, ongoing engagement. Looking at them together the datasets show that expansion hype depends on the platform, and is heavily tied to marketing milestones, suchs as launch trailers, and beta testing. 

### Limitations and Future Steps

Through my analysis I got clear insight into patterns of online engagement, but there were some limitations that shaped the results. </br>

*Google Trends* values are relative and not absolute. Trends platform scales from 0-100 based on the highest search volume within a certain timeframe. This means that the 100 for Dragonflight doesn't indicate the same number of searches for Midnight's 100. Since each keyword is measured against itself that limits the volume comparison across the three expansions. </br>

*YouTube* results were only based on 20 videos per expansion, to make the comparison even between the three. But that was only a small fraction of the videos, and the ones that were displayed were probably top ranked videos, this makes the dataset bias towards the top-performing content creators and official cinemetic releases. So the total and median view numbers may underrepresent smaller creator or regional content that could give us more insight into community engagement.</br> 

*Midnight* hasn't launched yet, so the data is still developing. Blizzard also hasn't hit some of its major marketing points- beta testing, launch trailers, and expansion guides. So the lower total view count and median performance should not be interpreted as a lack of interest, but rather as a look at Midnights early in its hype cycle. If I could have collected this data after its launch early 2026 I would have gotten a better comparison.</br> 

The engagement metrics don't directly reflect the players satisfaction or in-game behavior. The high searh interest or YouTube views may indicate curiosity, excitement, or controversy, but they don't reveal whether players are enjoying the expansion, purchasing or subscribing to the game, or if they are continuing to play it long-term.</br>

These analyses capture attention, not retention or approval, and are better understood if looked at in that context.</br> 

**Future Work**

- Comparing World of Warcrafs hype cylcles with other MMORPGs such as Final Fantasy XIV, Guild Wars 2, or Elder Scrolls Online. These games also produce spikes in attention around expansion reveals. Doing a cross-game comparison could help determine whether WoW's patterns are unique or part of a broader industry trend. </br>
- Tracking Midnights interest over a longer time period. Additional data could be captured thougout the beta phases, new trailers, patches, influencer coverage, etc. A follow-up analysis could reveal whether Midnight eventually surpasses earlier expansions in sustained engagement.

Overall while the current analysis successfully identifies cyclical patterns in online interest, these limitations and extensions highlight oppertunities to deepen the research and strengthen future findings.
